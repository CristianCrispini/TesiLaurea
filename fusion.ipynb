```python
from cv2 import imread, imwrite
from pyFusion.fusion import Fusion
from pyFusion.models import vgg16

import matplotlib.pyplot as plt
```

```python

# Parse arguments
images = ['images/MRI-SPECT/mr.png', 'images/MRI-SPECT/tc.png']
output = 'results/MRI-SPECT/'

# Read images
input_images = []
for image in images:
    input_images.append(imread(image))
```

```python

# Compute fusion image
model = vgg16.VGG16();
FU = Fusion(input_images, model)
fusion_img = FU.fuse()
```

```python

fig, (ax1, ax2, ax3) = plt.subplots(1, 3)

ax1.imshow(input_images[0])
ax2.imshow(input_images[1])
ax3.imshow(fusion_img)

imwrite('fused_lvl10.png', fusion_img)
```

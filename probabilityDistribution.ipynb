```python
from scipy.stats import entropy

def MI(fused,orig):
    """
    distance between a joint statistical distribution of two
    random variables X and Y from the case of being totally independent.
    In terms of entropies, and using Kvalseth normalisation
    """
    information = metrics.entropy_2d(fused, orig)
    Hx =metrics.entropy_2d(fused,fused)
    Hy =metrics.entropy_2d(orig,orig)

    return 2*(Hx + Hy - information) / (Hx + Hy)
    
```
